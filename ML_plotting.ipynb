```python
#matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)
print(x)
```

```python
print(y)
```

```python
print(z)
```

```python
#plotting the data
#sine wave]
plt.plot(x,y)
```

```python
plt.plot(x,z)
plt.show()
```

```python
#adding the x-axis and y-axis
plt.plot(x,y)
plt.xlabel('Angle')
plt.ylabel('sine value')
plt.title('sine wave')
plt.show()
```

```python
#parabola
x=np.linspace(-10,10,100)
y=x**2
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'r+')
plt.show()
```

```python
plt.plot(x,y,'r.')
plt.show()
```

```python
plt.plot(x,y,'gX')
plt.show()
```

```python
x=np.linspace(-5,5,50)
plt.plot(x,np.sin(x),'g+')
plt.plot(x,np.cos(x),'r--')
plt.show()
```

```python
#Bar plot
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
language=['Tamil','English','french','Telugu']
people=[100,120,200,300]
ax.bar(language,people)
plt.xlabel('Language')
plt.ylabel('people')
plt.title('language and people bar chart')
plt.show()
```

```python
#pie chart
fig1=plt.figure()
ax=fig1.add_axes([0,0,1,1])
language=['Tamil','English','french','Telugu']
people=[100,120,200,300]
ax.pie(people, labels=language, autopct="%1.1f%%")
plt.show()
```

```python
#scatter plot
x=np.linspace(0,10,30)
y=np.sin(x)
z=np.cos(x)
Fig2=plt.figure()
ax=Fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z, color='r')
plt.show()
```

```python
#3D scatter plot
fig3=plt.figure()
ax=plt.axes(projection='3d')
z=20*np.random.random(100)
x=np.sin(z)
y=np.cos(z)
ax.scatter(x,y,z,cmap='Blues')
plt.xlabel('Language')
plt.ylabel('people')
#plt.zlabel('comparison')
plt.title('language and people')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
#total bill vs tip dataset
tips=sns.load_dataset("tips")
```

```python
tips.head(10)
```

```python
sns.relplot(x="total_bill",y="tip", col='time',hue='smoker',style='smoker',size='size',data=tips)
```

```python
iris=sns.load_dataset("iris")
iris.head()
```

```python
sns.scatterplot(x='sepal_length',y="petal_length",hue="species",data=iris)
```

```python
sns.scatterplot(x='sepal_length',y="petal_width",hue="species",data=iris)
```
